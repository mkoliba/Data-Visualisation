```python
from csv import DictReader

the_reader = DictReader(open('data/201803_citibikenyc_tripdata 2.csv', 'r'))
data = []
for line_dict in the_reader:
    data.append(line_dict)
```

```python
data[0].keys()
```

```python
##get list of stations

stations = {}

for item in data:
    stations[item['start station name']] = {'latitude':  item['start station latitude'],'longitude': item['start station longitude']}
    stations[item['end station name']] = {'latitude':  item['end station latitude'],'longitude': item['end station longitude']}
    
print(stations)
```
